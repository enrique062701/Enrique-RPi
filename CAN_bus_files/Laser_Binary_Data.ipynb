```python
"""
So looking at the data, it seems that both the controller and the laser are communicating with each other and constantly echoing messages
on the CAN bus line. Since there are two atribution ids, it means that are two devices on the bus system. Must figure out which is which.
Once we figure out which is which, we can mimic the controller and send messages from the PiCAN hat.
"""
```

```python
Device_one = echo_message[echo_message['Atribration_id'] == '0x18820810']

Device_two = echo_message[echo_message['Atribration_id'] == '0x19000410']

print(len(Device_one))
print(len(Device_two))
```

```python
# So now getting all the unique messages from each device
device_one_data, device_one_dcounts = np.unique(Device_one['data'], return_counts = True)

print(device_one_data)
print(device_one_dcounts)
print(len(device_one_data))
```

```python
device_two_data, device_two_dcounts = np.unique(Device_two['data'], return_counts = True)

print(device_two_data)
print(device_two_dcounts)
print(len(device_two_data))
```
